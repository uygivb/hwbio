For the relation \\( R = \{(1,3),(2,2),(3,4)\}\\) on the set \\(\{1,2,3,4\}\\) the \\(R \cup (R^{-1}) ^{\ast} = \{(1,2),(2,2),(3,4),(3,1),(4,1),(4,3)\}\\) is not symmetric

```python
#4
%pylab inline
import networkx as nx

G1 = nx.Graph()
N = 10
G1 = nx.lollipop_graph(N,N-2)
G1.add_nodes_from(set(range(N+N-2, N+N-2 + N-1)))
for i in range(N-1):
    G1.add_edge(i, i + N+N-2)
nx.draw(G1, with_labels=True)
print("number of pendant verticles")
count = 0
for v in G1.nodes():
    if G1.degree(v) == 1:
        count+=1
print(count)
print("clique", nx.graph_clique_number(G1))
print("diameter", nx.diameter(G1))
```

```python
#5
G2 = nx.petersen_graph()
nx.draw_kamada_kawai(G2, with_labels=True)
print("if connected", nx.is_connected(G2))
print("if complete, 1st method", nx.density(G2) == 1)#For complete graph have to be 1
N = G2.number_of_nodes()
print(N)
print(G2.number_of_edges())
print("if complete, 2st method", G2.number_of_edges() == N*(N-1)/2) #checks if graph is complete
print("biggest cycle", nx.find_cycle(G2, source=None, orientation=None))
print("center vertices", nx.center(G2, e=None))
```
